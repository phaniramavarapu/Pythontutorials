```python
import pandas as pd
def convert_people_cell(cell):
    if cell == "n.a":
        return 'sam walton'
    return cell

def convert_eps_cell(cell):
    if cell == "not available":
        return None
    return cell

df = pd.read_excel("C:\\Users\\Phani\\Desktop\\Phani - New\\Kaggle Data Sets\\stock_data1.xlsx", "stock_data1", 
                  converters = {'people': convert_people_cell,
                               'eps': convert_eps_cell})
df
```

```python
df.to_excel("C:\\Users\\Phani\\Desktop\\Phani - New\\Kaggle Data Sets\\stock_data1.xlsx",
            sheet_name = "stock_data1", startrow=1, startcol=2, index = False)
```
